## Exploring graphs of the harmonic series

The harmonic series diverges, but very slowly.  This notebook show graphs of the partial sums and compares it to the related series with exponents greater than 1 and alternating exponents.

The first graph presents the partial sums of the series; the second compares these to the individual summands.  Then the summands are plotted by themselves. Then the series is plotted next to sum(1/n^k) where k > 1. These converging sums are then plotted on their own. The alternating harmonic series is then plotted. Finally the alternating converging sums are also plotted

```python
import matplotlib.pyplot as plt
import math
```

```python
#this variable will determine the n to sum to 

up_to_n = 100
```

```python
#sum up to n function
def harmonic_sum(n):
    #returns sum up to n
    return sum([1/x for x in range(1, n+1)])
```

```python
#plot the partial sums
plt.figure(figsize=(10,10))

x = range(1,up_to_n)
y = [harmonic_sum(n) for n in x]

plt.scatter(x,y,s=1)
```

```python
#plot the summands next to the cumulative sum, and the line y=1 for comparison
plt.figure(figsize=(10,10))

x = range(1,up_to_n)
y1 = [1/x for x in x]
line = [1 for x in x]

plt.scatter(x, y, s=1)
plt.scatter(x, y1, s=1)
plt.plot(x, line, c='red')
```

```python
#plot the summands alone
plt.figure(figsize=(10,10))

x = range(1,up_to_n)
y1 = [1/x for x in x]

plt.scatter(x, y1, s=1)
```

```python
#now compare to other sums of the form 1/n^x where x is a positive integer

#first make a function to return the partial sums up to n with exponent x
def one_over_n_sum(n, x):
    #returns sum up to n
    return sum([1/(k**x) for k in range(1, n+1)])
```

```python
#make a series of plots
exponents = [e for e in range(1,10)]

#make a list of the partial sums
sums_to_plot = []
for val in exponents:
    sums_to_plot.append([one_over_n_sum(n, val) for n in range(1,up_to_n)])
```

```python
#now plot the partial sums
plt.figure(figsize=(10,10))
for f in sums_to_plot:
    x = range(1,up_to_n)
    plt.plot(x, f)
plt.show()
#can clearly see that exponents > 1 converge, and the clear divergence of the 
```

```python
#plot just the converging series
exponents = [e for e in range(2,10)]

#make a list of the partial sums
sums_to_plot = []
for val in exponents:
    sums_to_plot.append([one_over_n_sum(n, val) for n in range(1,up_to_n)])
plt.figure(figsize=(10,10))
for f in sums_to_plot:
    x = range(1,up_to_n)
    plt.plot(x, f)
plt.show()
```

```python
#plot the alternating harmonic sum to see the convergence 

def alternating_sum(n,x):
    #plot an alternating sum up n with exponent x
    return sum([(((-1)**(k+1))/(k**x)) for k in range(1, n+1)])

y_alt_sum = [alternating_sum(n,1) for n in range(1, up_to_n)]

y_log_2 = [math.log(2) for n in range(1, up_to_n)]

#now plot
plt.figure(figsize=(10,10))

plt.scatter(x,y_alt_sum,s=1,c='red')
plt.plot(x, y_log_2) #to compare
#plt.xlim(800,1000)
plt.ylim(0.6,0.8)
#clearly converges to 0.7 ~ ln(2) as expected
```

```python
#also plot the alternating sequences for the converging series
def one_over_n_sum_alt(n, x):
    #returns sum up to n
    return sum([((-1)**(k+1))/(k**x) for k in range(1, n+1)])

exponents = [e for e in range(2,10)]

#make a list of the partial sums
sums_to_plot = []
for val in exponents:
    sums_to_plot.append([one_over_n_sum_alt(n, val) for n in range(1,up_to_n)])
plt.figure(figsize=(10,10))
for f in sums_to_plot:
    x = range(1,up_to_n)
    plt.scatter(x, f, s=1)
plt.show()
```
